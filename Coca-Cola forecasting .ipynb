### SIMPLE EXPONENTIAL SMOOTHING

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing,ExponentialSmoothing
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
```

```python
model_simple=SimpleExpSmoothing(train_data["Sales"]).fit()
simple_test_predict=model_simple.predict(start=train_len,end=train_len+test_len)
mse_simple=mean_squared_error(test_data['Sales'],simple_test_predict)
rmse_simple=np.sqrt(mse_simple)
mae_simple=mean_absolute_error(test_data['Sales'],simple_test_predict)
mape_simple=mean_absolute_percentage_error(test_data['Sales'],simple_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
simple_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_simple.resid.plot(label='Resid')
plt.legend()
```

### DOUBLE EXPONENTIAL SMOOTHING WITH MULTIPLICATIVE TREND

```python
model_double=ExponentialSmoothing(train_data['Sales'],trend='mul').fit()
double_mul_test_predict=model_double.predict(start=train_len,end=train_len+test_len)
mse_double_mul=mean_squared_error(test_data['Sales'],double_mul_test_predict)
rmse_double_mul=np.sqrt(mse_double_mul)
mae_double_mul=mean_absolute_error(test_data['Sales'],double_mul_test_predict)
mape_double_mul=mean_absolute_percentage_error(test_data['Sales'],double_mul_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
double_mul_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_double.resid.plot(label='Resid')
plt.legend()
```

```python
train=df1.iloc[:31]
test=df1.iloc[31:]
```

### LINEAR MODEL

```python
import statsmodels.formula.api as smf
```

```python
model_linear=smf.ols('Sales~t',data=train).fit()
linear_test_predict=pd.Series(model_linear.predict(pd.DataFrame(test['t'])))
mse_linear=mean_squared_error(test['Sales'],linear_test_predict)
rmse_linear=np.sqrt(mse_linear)
mae_linear=mean_absolute_error(test['Sales'],linear_test_predict)
mape_linear=mean_absolute_percentage_error(test['Sales'],linear_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
linear_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_linear.resid.plot(label='Resid')
plt.legend()
```

### EXPONENTIAL MODEL

```python
model_exp=smf.ols('log_sales~t',data=train).fit()
exp_test_predict=pd.Series(model_exp.predict(pd.DataFrame(test['t'])))
mse_exp=mean_squared_error(test['Sales'],exp_test_predict)
rmse_exp=np.sqrt(mse_exp)
mae_exp=mean_absolute_error(test['Sales'],exp_test_predict)
mape_exp=mean_absolute_percentage_error(test['Sales'],exp_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
exp_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_exp.resid.plot(label='Resid')
plt.legend()
```

### QUADRATIC MODEL

```python
model_qua=smf.ols('Sales~t+t_squared',data=train).fit()
qua_test_predict=pd.Series(model_qua.predict(pd.DataFrame(test[['t','t_squared']])))
mse_qua=mean_squared_error(test['Sales'],qua_test_predict)
rmse_qua=np.sqrt(mse_qua)
mae_qua=mean_absolute_error(test['Sales'],qua_test_predict)
mape_qua=mean_absolute_percentage_error(test['Sales'],qua_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
qua_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_qua.resid.plot(label='Resid')
plt.legend()
```

### ADDITIVE SEASONALITY MODEL

```python
model_add=smf.ols('Sales~Q1+Q2+Q3',data=train).fit()
add_test_predict=pd.Series(model_add.predict(pd.DataFrame(test[['Q1','Q2','Q3']])))
mse_add=mean_squared_error(test['Sales'],add_test_predict)
rmse_add=np.sqrt(mse_add)
mae_add=mean_absolute_error(test['Sales'],add_test_predict)
mape_add=mean_absolute_percentage_error(test['Sales'],add_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
add_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_add.resid.plot(label='Resid')
plt.legend()
```

### ADDITIVE QUADRATIC MODEL

```python
model_add1=smf.ols('Sales~t+t_squared+Q1+Q2+Q3',data=train).fit()
add1_test_predict=pd.Series(model_add1.predict(pd.DataFrame(test[['Q1','Q2','Q3','t','t_squared']])))
mse_add1=mean_squared_error(test['Sales'],add1_test_predict)
rmse_add1=np.sqrt(mse_add1)
mae_add1=mean_absolute_error(test['Sales'],add1_test_predict)
mape_add1=mean_absolute_percentage_error(test['Sales'],add1_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
add1_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_add1.resid.plot(label='Resid')
plt.legend()
```

### MULTIPLICATIVE SEASONALITY MODEL

```python
model_mu=smf.ols('log_sales~Q1+Q2+Q3',data=train).fit()
mu_test_predict=pd.Series(model_mu.predict(pd.DataFrame(test[['Q1','Q2','Q3']])))
mse_mu=mean_squared_error(test['Sales'],mu_test_predict)
mae_mu=mean_absolute_error(test['Sales'],mu_test_predict)
rmse_mu=np.sqrt(mse_mu)
mape_mu=mean_absolute_percentage_error(test['Sales'],mu_test_predict)
plt.figure(figsize=(20,6))
df['Sales'].plot(label='original data',marker='o')
mu_test_predict.plot(label='predict data')
plt.legend()
plt.show()
plt.figure(figsize=(20,6))
model_mu.resid.plot(label='Resid')
plt.legend()
```
